```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## a) Plot the Count of Famous Food Outlets / Restaurants in Bengaluru

```python
zomato_df=pd.read_csv("E:/Data/zomato_restaurants.csv")
zomato_df
```

```python
# zomato_df['name'].value_counts()
zomato_df['name'].value_counts().hist(bins = 15, figsize=(35,10),color = 'red')
plt.title("Histogram of Famous food outlets", fontsize = 16)
plt.xlabel("Famous food outlets")
plt.ylabel("Frequency(in Range)")
plt.show()
```

## b) Display accepting online orders vs not accepting online orders count % distribution

```python
plt.figure(figsize=(8,8))
plt.pie(zomato_df['book_table'].value_counts(),labels=zomato_df['book_table'].unique(),autopct='%1.2f%%',explode=[0,0])
plt.legend()
plt.show()
```

## c) Find the popular restaurant type in Bengaluru. 


```python
plt.figure(figsize=(16,10))
restaurant=sns.countplot(zomato_df['rest_type'])
plt.xticks(rotation=90)
```

## d)	Display accepting Table Booking vs Not accepting table booking count % distribution

```python
plt.figure(figsize=(8,8))
plt.pie(zomato_df['book_table'].value_counts(),labels=zomato_df['online_order'].unique(),autopct='%1.2f%%',explode=[0,0])
plt.legend()
plt.show()
```

## e) Histogram and Box Plot of Approx Cost Distribution. Note: Replace ‘,’ from the values using replace(), lambda function with replace()

```python
# histogram
zomato_df['approx_cost(for two people)'].hist(bins = 15, figsize=(35,10),color = 'pink')
plt.title("Histograph Plot of Approx_cost", fontsize = 12)
plt.xlabel("approx_cost(for two people)")
plt.ylabel("Frequency(in Range)")
plt.show()
```

```python
#boxplot
zomat=zomato_df['approx_cost(for two people)'].str.replace(",","")
plt.figure(figsize=(11,8))
sns.boxplot(data=zomat,palette='plasma')
```

## f) Histogram and Box Plot of Restaurant Rating Distribution. Note: Replace ‘,’ from the values using replace(), lambda function with replace()

```python
zomato_df['rate'].hist(bins = 15, figsize=(35,10),color = 'orange')
plt.title("Histograph Plot of Rate", fontsize = 12)
plt.xlabel("Rate")
plt.ylabel("Frequency(in Range)")
plt.show()
```

## g) Rate vs Approx Cost (for two people). Also categorise them into online_order available or not. 

```python
 
sns.set_style('whitegrid')
plt.figure(figsize=(20,17))
sns.scatterplot(x = "rate", y = "approx_cost(for two people)", data = zomato_df, palette='plasma', hue = 'online_order')
plt.title('Cost vs Rating comparison')
plt.xlabel('Rating between 0 to 5')
plt.ylabel('Approx. cost for 2 people')
```

## h) Distribution of Approx Cost for two people. Is it skewed?

```python
zomato_1=zomato_df['approx_cost(for two people)'].str.replace(",", "")
plt.figure(figsize=(6,6))
sns.set_style('darkgrid')
sns.distplot(zomato_1, color= 'green')
plt.title('Approx. cost for two people in city of Banguluru')
plt.xlabel('Cost for two people')
```

## i)	Is there any difference b/w votes of restaurants accepting and not accepting online orders?

```python
sns.set_style('whitegrid')
plt.figure(figsize=(9,6))
ax = sns.barplot(x = "online_order", y = "votes", data = zomato_df, ci = None, palette='plasma')
```

## j)	Which are the Top 20 most common restaurant types in Bengaluru?

```python
plt.figure(figsize=(10,7))
sns.set_style('darkgrid')
zomato_1=zomato_df['rest_type'].value_counts()[:20]
sns.barplot(x=zomato_1,y=zomato_1.index,palette='dark')
sns.despine
plt.title("Top 20 restaurants in Bengaluru", fontsize = 12)
plt.xlabel("counts")
plt.ylabel("type of restaurant")
```
